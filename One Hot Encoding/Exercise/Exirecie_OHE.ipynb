```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("carprices.csv")
df
```

```python
plt.scatter(df['Sell Price($)'],df['Age(yrs)'])
```

```python
dummies = pd.get_dummies(df['Car Model'], dtype = int)
dummies
```

```python
marged = pd.concat([df,dummies], axis='columns')
marged
```

```python
final = marged.drop(['Mercedez Benz C class','Car Model'], axis='columns')
final
```

```python
x = final.drop('Sell Price($)',axis='columns')
x
```

```python
y = final['Sell Price($)']
y
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

## Price of mercedez benz 4yr old with mileage 45000

```python
model.predict([[45000,4,0,0]])
```

## Price of BMW X5 that is 7 yr old with mileage 86000

```python
model.predict([[86000,7,0,1]])
```
