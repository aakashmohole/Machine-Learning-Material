```python
animalList2 = {
    'Animal':['Cat','Dog','Tiger','Monkey','Bear','Rabit'],
    'Age' : [3,5,np.nan,3,5,7],
    'Visits': [1,np.nan,1,1,1,np.nan],
    'Suitable':['Yes','No',np.nan,'Yes','No','No']
}

srNo = [1,2,3,4,5,6]

df4 = pd.DataFrame(animalList2, index = srNo)
df4
```
