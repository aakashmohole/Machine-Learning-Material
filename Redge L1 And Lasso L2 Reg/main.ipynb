```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("melborn.csv")
df.head()
```

```python
df.shape
```

```python
df.nunique()
```

```python
drop_cl = ["Address", "Date", "Postcode","YearBuilt","Lattitude", "Longtitude"]
df = df.drop(drop_cl, axis="columns")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
fill_cl = ["Distance", "Bedroom2", "Bathroom", "Car", "Propertycount"]
df[fill_cl] = df[fill_cl].fillna(0)
df.isna().sum()
```

```python
df["Landsize"] = df["Landsize"].fillna(df.Landsize.mean())
df["BuildingArea"] = df["BuildingArea"].fillna(df.BuildingArea.mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df = pd.get_dummies(df, drop_first=True)

# Convert True/False values to 0/1 integers
df = df.astype(int)
df.head()
```

```python
x = df.drop("Price",  axis=1)
y = df["Price"]
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(train_x, train_y)
```

## Cheaking Model Score 

```python
reg.score(test_x,test_y)
```

## Model is Overfit

```python
reg.score(train_x, train_y)
```

## Import model Lasso L1

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha = 50, max_iter= 100, tol=0.1)
lasso_reg.fit(train_x, train_y)
```

```python
lasso_reg.score(test_x, test_y)
```

```python
lasso_reg.score(train_x, train_y)
```

## Import Model Ridge L2 

```python
from sklearn import linear_model
ridge_reg = linear_model.Ridge(alpha = 50, max_iter= 100, tol=0.1)
ridge_reg.fit(train_x, train_y)
```

```python
ridge_reg.score(train_x,train_y)
```

```python
ridge_reg.score(test_x, test_y)
```
