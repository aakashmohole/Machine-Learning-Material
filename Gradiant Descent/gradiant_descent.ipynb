```python
import numpy as np

def gradiant_descent(x, y):
    m_curv = b_curv = 0
    itreations = 1000
    n = len(x)
    learning_rate = 0.8
    
    for i in range(itreations):
        y_pre = m_curv * x + b_curv
        cost = (1/n) * sum([val**2 for val in (y-y_pre)])
        md = -(2/n)*sum(x*(y-y_pre))
        bd = -(2/n)*sum(y-y_pre)
       
        m_curv = m_curv - learning_rate * md
        b_curv = b_curv - learning_rate * bd
        
        print ("m {}, b {}, cost {} iteration {}".format(m_curv,b_curv,cost, i))


x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradiant_descent(x,y)
```
