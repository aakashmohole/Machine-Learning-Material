# Load Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# resize the graph
plt.figure(figsize=(8,3),dpi = 200)



#color used to give color in line
#line width used to give the width of line
#marker are dots in line and markersize is the size of that dots

# plt.plot(x,y, label='2x', color = 'red', linewidth = 3, linestyle= '--', marker ='.', markersize = 10,markeredgecolor= 'blue')

# short hand notation
# fmt = '[color],marker],line]]'
plt.plot(x,y, 'r.--', label = '2x')

### line number two

x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2, 'g', label = '2x')
plt.plot(x2[5:],x2[5:] ** 2, 'g--', label = '2x') # change the half line


plt.title("First Graph", fontdict={'fontname' : 'sans-serif', 'fontsize' : 20}) #title for the graph
plt.xlabel("X-axis", ) #title for x axis
plt.ylabel("Y-axis") #title for y axis

plt.xticks([0,1,2,3,4]) #arrange the points on axis decimal to intger
# plt.yticks([0,2,4,6,6.5,8])

#savethe graph
plt.savefig('mygraph.png',dpi = 300)
plt.show()
```
