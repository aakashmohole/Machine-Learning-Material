```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

# Line Plot

```python
var1 = [1,2,3,5,6,7]
var2 = [2,3,8,1,6,7]

data = pd.DataFrame({"var1" : var1, "var2" : var2})
sns.lineplot(x="var1" , y="var2", data = data)
plt.show()
```

```python
df = sns.load_dataset("penguins")
df
```

```python
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = df.head(100), hue = "sex", style="sex")
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = df.head(100), hue = "sex", style="sex", palette = "Accent")
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = df.head(100), hue = "sex", style="sex", palette = "Accent",
            dashes=False)
plt.show()
```

# Bar Plot

```python
df.head()
```

```python
sns.barplot(x = "island", y="bill_length_mm",data = df, hue= "sex")
plt.show()
```

```python
ordr = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y="bill_length_mm",data = df, hue= "sex", order = ordr )
plt.show()
```

```python
ordr = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y="bill_length_mm",data = df, hue= "sex", order = ordr, hue_order=["Female", "Male"] )
plt.show()
```

```python
ordr = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y="bill_length_mm",data = df, hue= "sex", order = ordr, ci = 10, n_boot=2)
plt.show()
```

```python
ordr = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y="bill_length_mm",data = df, hue= "sex", order = ordr, orient="v", palette = "icefire" )
plt.show()
```

```python
ordr = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y="bill_length_mm",data = df, hue= "sex", order = ordr, saturation=10 )
plt.show()
```

```python
ordr = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y="bill_length_mm",data = df, hue= "sex", order = ordr, saturation=10, errcolor="b" )
plt.show()
```

```python
ordr = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y="bill_length_mm",data = df, hue= "sex", order = ordr, saturation=10, errcolor="r", capsize=0.7 )
plt.show()
```
